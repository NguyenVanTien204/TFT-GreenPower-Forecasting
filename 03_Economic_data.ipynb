```python
import pandas as pd
from fredapi import Fred
import matplotlib.pyplot as plt

# 1. Cấu hình API
api_key = '87841008dbb437a68c2481c04e8a2738'
fred = Fred(api_key=api_key)

# 2. Danh sách các mã có thể chứa dữ liệu IIP Việt Nam (Dự phòng)
# PRINTO01VNM661N: Sản xuất công nghiệp tổng thể (OECD)
possible_ids = ['PRINTO01VNM661N', 'VNMIPPRODMSMEI']

df = None

print("Đang tìm kiếm dữ liệu trên FRED...")

for s_id in possible_ids:
    try:
        data = fred.get_series(s_id, observation_start='2019-01-01')
        if not data.empty:
            df = pd.DataFrame(data, columns=['IIP_Vietnam'])
            print(f"--- Đã tìm thấy dữ liệu với mã: {s_id} ---")
            break
    except Exception:
        continue

if df is not None:
    # 3. Tiền xử lý
    df.index.name = 'Date'
    df = df.dropna()
    
    # 4. Hiển thị kết quả
    print("\n5 tháng gần nhất:")
    print(df.tail())

    # 5. Vẽ biểu đồ
    plt.figure(figsize=(12, 6))
    plt.plot(df.index, df['IIP_Vietnam'], color='red', marker='s', markersize=3, label='IIP Vietnam')
    plt.title('Chỉ số Sản xuất Công nghiệp (IIP) Việt Nam (2019 - 2026)', fontsize=14)
    plt.ylabel('Chỉ số (Index)')
    plt.legend()
    plt.grid(True, linestyle=':', alpha=0.6)
    plt.show()
else:
    print("Không tìm thấy mã nào hoạt động. Đang thực hiện tìm kiếm thủ công...")
    # Tính năng tìm kiếm: liệt kê các series liên quan đến Vietnam Industrial
    search_results = fred.search('Vietnam Industrial Production')
    print("Gợi ý các mã series có sẵn trên FRED hiện tại:")
    print(search_results[['title', 'id']].head(10))
```
