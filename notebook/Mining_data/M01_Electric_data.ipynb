```python
import pandas as pd
import polars as pl
import numpy as np
import matplotlib.pyplot as plt
```

```python
monthly_data = pd.read_csv('data\\vietnam_data\\monthly_electric_data.csv')
yearly_data = pd.read_csv('data\\vietnam_data\\yearly_electric_data.csv')
```

```python
monthly_data.head()
```

```python
vietnam_monthly_data = monthly_data[monthly_data['Area'] == 'Viet Nam']
vietnam_monthly_data.head()
```

```python
# fillter data
vietnam_data = vietnam_monthly_data.drop(columns=['Area', 'ISO 3 code', 'Area type', 'Continent', 'Ember region', 'EU', 'OECD', 'G20', 'G7', 'ASEAN'])
vietnam_data.head()
```

```python
vietnam_data.tail()
```

```python
vietnam_data.info()
```

```python
vietnam_data['Date'] = pd.to_datetime(vietnam_data['Date'], format='%Y-%m-%d')
vietnam_data.info()
```

```python
# 1. Chỉ lọc lấy đơn vị TWh để phân tích sản lượng
df_twh = vietnam_data[vietnam_data['Unit'] == 'TWh'].copy()

# 2. Pivot dữ liệu: Chuyển Variable thành cột
df_wide = df_twh.pivot_table(index='Date', 
                             columns='Variable', 
                             values='Value').reset_index()

# 3. Điền các giá trị thiếu bằng 0 (ví dụ Solar/Wind năm 2019 có thể chưa có dữ liệu)
df_wide = df_wide.fillna(0)
```
